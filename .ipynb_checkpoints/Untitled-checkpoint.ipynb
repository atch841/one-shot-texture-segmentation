```python
all_texture = []
img_folder_p = '/fast_data/texture/dtd/dtd-r1.0.1/dtd/images/'
for img_folder_c in os.listdir(img_folder_p):
    img_folder = img_folder_p + img_folder_c + '/'
    for img_name in os.listdir(img_folder):
        img = cv2.imread(img_folder + img_name)
        if type(img) == type(None):
            print(img_folder + img_name, 'is not a image.')
            continue
        if img.shape[0] < 256 or img.shape[1] < 256:
            print(img_folder + img_name, 'is too small.')
            continue
        mid_0 = int(img.shape[0] / 2)
        mid_1 = int(img.shape[1] / 2)
        img = img[mid_0-128:mid_0+128,mid_1-128:mid_1+128,:]
        all_texture.append(img)
all_texture = np.array(all_texture)
all_texture.shape
```

```python
from tensorflow.contrib import autograph
```

```python
@autograph.convert()
def check_time(a):
    k = 0
    while k < 256*256:
        i = k // 256
        j = k % 256
        a += i + j
        print(a)
        k += 1
```

```python
with tf.Session() as sess:
    sess.run(check_time(tf.constant(0.0)))
```

```python
@autograph.convert()
def weighted_cross_entropy(y_pred, y_gt):
	l = tf.reduce_sum(tf.multiply(y_gt, tf.log(y_pred)), axis=1)
	l = tf.reduce_sum(l, axis=1)
	l = tf.reduce_sum(l, axis=1)
	n = tf.reduce_sum(y_gt, axis=1)
	n = tf.reduce_sum(n, axis=1)
	n = tf.reduce_sum(n, axis=1)
	l = -1 * tf.reduce_sum(l / n)
	return l

@autograph.convert()
def total_loss(y_pred, y_gt):
	return tf.reduce_sum(weighted_cross_entropy(y_pred, y_gt) + weighted_cross_entropy((1 - y_pred), (1 - y_gt)))
```

```python
a = tf.convert_to_tensor(np.random.random((5, 256, 256, 1)), dtype=tf.float32)
b = tf.convert_to_tensor(np.random.random((5, 256, 256, 1)), dtype=tf.float32)
```

```python
l = total_loss(a, b)
```
